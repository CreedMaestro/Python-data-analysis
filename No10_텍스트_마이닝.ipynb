```python
# 대통령 연설문 로드
moon = open('speech_moon.txt', encoding = 'utf-8').read()
moon
```

```python
# 불필요한 문자 제거
import re
moon = re.sub('[^가-힣]', ' ', moon)  #[^가-힣]은 '한글이 아닌 모든 문자'를 의미하는 정규표현식.
moon
```

```python
# 연설문에서 명사만 추출
# hannanum 할당
import konlpy
hannanum = konlpy.tag.Hannanum()

# 명사 추출
hannanum.nouns('대한민국의 영토는 한반도와 그 부속도서로 한다')
```

```python
# 연설문에서 명사 추출
nouns = hannanum.nouns(moon)
nouns
```

```python
# 데이터 프레임으로 변환
import pandas as pd
df_word = pd.DataFrame({'word' : nouns})
df_word
```

```python
# 단어 빈도표 생성
```

```python
# 글자 수 추가

df_word['count'] = df_word['word'].str.len()
df_word
```

```python
# 두 글자 이상 단어만 남기기
df_word = df_word.query('count >= 2')
df_word.sort_values('count')
```

```python
# 단어 빈도 구하기
# 단어별 그룹핑하여 빈도 산출 후 내림차순 정렬

df_word = df_word.groupby('word', as_index = False) \
                  .agg(n = ('word', 'count')) \
                  .sort_values('n', ascending = False)
                
df_word
```

```python
# 빈도 상위 20개 단어 추출
top20 = df_word.head(20)
top20
```

```python
# 막대 그래프 생성
import seaborn as sns
import matplotlib.pyplot as plt

plt.rcParams.update({'font.family'    : 'NanumBarunGothic',  # 한글 폰트 설정
                     'figure.dpi'     : '120',            # 해상도 설정
                     'figure.figsize' : [6.5, 6]})        # 가로 세로 크기 설정

# 막대 그래프 만들기
sns.barplot(data = top20, y = 'word', x = 'n')
```

# 워드 클라우드 생성성

```python
# 워드클라우드 패키지 설치
!pip install wordcloud
```

```python
# 한글 폰트 설정
font = 'NanumBarunGothic'
```

```python
# 데이터 프레임 딕셔너리 타입으로 변환
dic_word = df_word.set_index('word').to_dict()['n']
dic_word
```

```python
# 워드클라우드 뭉치 생성

from wordcloud import WordCloud
wc = WordCloud(random_state = 1234,         # 난수 고정
               font_path = font,            # 폰트 설정
               width = 400,                 # 가로 크기
               height = 400,                # 세로 크기
               background_color = 'white')  # 배경색
```

```python
# 워드 클라우드 생성
gen = wc.generate_from_frequencies(dic_word)

# 워드 클라우드 출력하기
plt.figure(figsize = (10, 10))  # 가로, 세로 크기 설정
plt.axis('off')                 # 테두리 선 없애기
plt.imshow(gen)       # 워드 클라우드 출력 
```

```python
# 워드클라우드 구름 모양 변경위해 이미지 로드드

import PIL
icon = PIL.Image.open('cloud.png')
```

```python
# 로드한 이미지 파일로 구름 모양 변경

import numpy as np
img = PIL.Image.new('RGB', icon.size, (255, 255, 255))
img.paste(icon, icon)
img = np.array(img)
```

# 기사 댓글 텍스트 마이닝

```python
# 데이터 로드
df = pd.read_csv('news_comment_BTS.csv', encoding = 'UTF-8')

# 데이터 살펴보기
df.info()
```

```python
#  댓글 변수인 reply 불필요한 문자 제거 후 한글만 남겨 결과물 확인

df['reply'] = df['reply'].str.replace('[^가-힣]', ' ', regex = True)
df['reply'].head()
```

```python
# 꼬꼬마(Kkma) 형태소 분석기 이용해 댓글에서 명사 추출
# Kkma는 띄어쓰기 오류가 있는 문장에서도 형태소를 잘 추출하는 장점이 있어 댓글처럼 비정제 텍스트 분석에 용이
# KKma 생성
Kkma = konlpy.tag.Kkma()
```

```python
# reply는 데이터프레임에 들어있는 변수라서 Kkma.nouns()에 적용 불가. apply()활용하여 함수가 각행의 값을 따로 처리하도록 지정
nouns = df['reply'].apply(Kkma.nouns)
nouns
```

```python
# 단어 빈도표 생성
# nouns()는 행마다 여러 단어가 리스트 자료 구조로 표시되어 df.explode()를 이용해 한 행에 한 단어만 출력되도록 변경

nouns = nouns.explode()
nouns
```

```python
# nouns 데이터 프레임 생성
df_word = pd.DataFrame({'word' : nouns})

# 글자 수 추가
df_word['count'] = df_word['word'].str.len()

# 두 글자 이상 단어만 남기기
df_word = df_word.query('count >= 2')
df_word
```

```python
# 단어로 그룹핑하여 빈도표 생성후 내림차순 정렬

df_word = df_word.groupby('word', as_index = False) \
                  .agg(n = ('word', 'count')) \
                  .sort_values('n', ascending = False)

df_word
```

```python
# 단어 빈도 막대 그래프 생성 전 고빈도 상위 20개 추출출

top20 = df_word.head(20)
top20
```

```python
# 그래프 가로 세로 크기 설정

plt.rcParams.update({'figure.figsize': [6.5, 6]})

# 막대 그래프 생성
sns.barplot(data = top20, y ='word', x = 'n')
```

```python
# 데이터 프레임 딕셔너리로 변환
dic_word = df_word.set_index('word').to_dict()['n']
```

```python
# 워드클라우드 뭉치 만들기

wc = WordCloud(random_state = 1234,         # 난수 고정
               font_path = font,            # 폰트 설정
               width = 400,                 # 가로 크기
               height = 400,                # 세로 크기
               background_color = 'white',  # 배경색
               mask = img)                  # mask 설정
```

```python
# 워드 클라우드 생성
img_wordcloud = wc.generate_from_frequencies(dic_word)

# 워드 클라우드 출력
plt.figure(figsize = (10, 10))  # 가로, 세로 크기 설정
plt.axis('off')                 # 테두리 선 없애기
plt.imshow(img_wordcloud)       # 워드 클라우드 출력
```
