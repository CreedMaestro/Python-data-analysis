```python
# 모든 변수 요약 통계량 한번에 구하기 위해선 agg대신 groupby 뒤에 산술함수 입력
exam.groupby('nclass').mean()
```

```python
# 중간고사 데이터 생성
test1 = pd.DataFrame({'id'     : [1, 2, 3, 4, 5],
                     'midterm' : [60, 80, 70, 90, 85]})

#기말고사 데이터 생성
test2 = pd.DataFrame({'id'     : [1, 2, 3, 4, 5],
                     'final'   : [70, 83, 65, 95, 80]})
```

```python
test1
```

```python
test2
```

```python
# 데이터를 가로로 합칠 때. [열 병합] pd.merge() 함수 활용
# 오른쪽 데이터 프레임을 왼쪽 데이터 프레임에 결합하도록 how = 'left' 입력
# 데이터 합칠 기준 on = '' 입력
```

```python
# id 기준으로 합쳐서 total에 할당
total = pd.merge(test1, test2, how = 'left', on = 'id')
total
```

```python
# 반별 담임교사 명단 데이터 생성
name = pd.DataFrame({'nclass'   : [1, 2, 3, 4, 5],
                     'teacher'  : ['kim', 'lee', 'park', 'choi', 'jung']})
name
```

```python
# 세로로 합칠 때 [행 병합]은 pd.concat() 함수 사용
# 1 ~5 번 학생 데이터 만들기
group_a = pd.DataFrame({'id'     : [1, 2, 3, 4, 5],
                       'test'    : [60, 80, 70, 90, 85]})

# 6~ 10번 학생 데이터 만들기
group_b = pd.DataFrame({'id'     : [6, 7, 8, 9, 10],
                       'test'    : [70, 83, 65, 95, 80]})
```

```python
group_a
```

```python
group_b
```

```python
# 행병합. 인덱스 중복 방지 위해 ignore_index 지정. concat는 데이터 변수명이 같아야 하며, 다를 경우 pd.rename()함수로 변경후 병합.
group_all = pd.concat([group_a, group_b], ignore_index = True)
group_all
```

```python
midwest = pd.read_csv('midwest.csv')
```

```python
midwest.info()
```

```python
# 전체 인구 대비 미성년 인구 백분율
midwest['popchild'] = (midwest['poptotal']-midwest['popadults'])/midwest['poptotal'] * 100
```

```python
# 미성년 인구 백분율이 가장 높은 상위 5개 county의 미성년 인구 백분율 출력
midwest.sort_values('popchild', ascending = False).head() \
        [['county', 'popchild']]
```

```python
# 미성년 비율 등급 변수 추가, 각 등급에 몇 개의 지역이 있는지 확인
midwest['childratio'] = np.where(midwest['popchild'] >= 40, 'large', 
                                 np.where(midwest['popchild'] >= 30, 'middle', 'small'))
```

```python
midwest.groupby('childratio') \
        .agg(n = ('childratio', 'count'))
```

```python
# 전체 인구 대비 아시아인 인구 백분율 변수 추가 후 하위 10개 지역의 state, county, 아시아인 인구 백분율 출력
midwest['asianratio'] = (midwest['popasian'] / midwest['poptotal']) * 100
```

```python
midwest[['state', 'county', 'asianratio']] \
        .sort_values('asianratio').head(10)
```
